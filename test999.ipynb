```python
# iii    C:\Users\Administrator\Desktop\project 3
import csv 
import numpy as np
import pandas as pd
with open (r'C:\Users\Administrator\Desktop\project 3\tournamentinfo.txt') as txtfile:
    readCSV = csv.reader(txtfile)
    table=[]
    for row in readCSV:
        table.append(row)
table   
```

```python
#remove the first three unmeaningful rows, the rows after the forth can seem like a regular formulas
table1 = table[3:]
table1
```

```python
length = len(table1)
print(length)

b = np.arange(1,length,3)
row = 0
table2 = []
for a in table1:
    if (row in b):
        table2.append(a)
    row = row+1
table2
```

```python
length = len(table1)
c = np.arange(2,length,3)
row = 0
table3 = []
for a in table1:
    if (row in c):
        table3.append(a)
    row = row+1
table3
```

```python
import re
str1 = re.compile('[A-Z]+\s[A-Z]+\s[A-Z]*')
name = str1.findall(str(table2))
name
```

```python
str2 = re.compile('[A-Z]{2}')
state = str2.findall(str(table3))
state
```

```python
str3 = re.compile('[0-9]\.[0-9]')
rate = str3.findall(str(table2))
rate
```

```python
str4 = re.compile('[A-Z]{1}\:\s+[0-9]+')
pre1 = str4.findall(str(table3))
pre1
```

```python
str5 = re.compile('[0-9]+')
pre2 = str5.findall(str(pre1))
pre2
```

```python
# There might be some information missing or some of them have no opponent number in 40 set of data, and those data start with"H","U","B" or"X",
# so we pick them out.   

str10 = re.compile('[A-Z]{1}\s+[0-9]{1,2}|\|[H,U,B,X]\s+')
opponents10 = str10.findall(str(table2))
opponents10
```

```python
str11 = re.compile('[0-9]+|\s{4,}')
opponents11 = str11.findall(str(opponents10))
opponents11
```

```python
# change '    ' which with 4 spaces to '64', later will add the 65th number in opponents11 list a value of 0;
pre3=['64' if x == '    ' else x for x in opponents11]
pre3
```

```python
# change str to int
pre4 = list(map(int, pre3))
print(pre4)
```

```python
# change pre2 str to int
pre2_int = list(map(int, pre2))
print(pre2_int)
```

```python
# add the 65th number of 'pre2_int' a value of 0
pre4 = pre2_int.insert(64,0) 
print(pre4)
```
