```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/YALINYAN-YU/DAV5400/master/diamonds.csv")
df
```

1. What proportion of diamonds are between .30 and 1.08 carats?
We need to extract the carats are between .30 and 1.08, to get the total number of the ones meet this requirments. 
The propertion is this number divid the total number of dimonds. 

```python
#step1,to extract the carats are between .30 and 1.08, and then count the total rows of carats which are between .30 and 1.08

carats = df[(df['carat']>0.3) & (df['carat']<1.08)]
carats = carats .shape[0]
carats
```

```python
#step2,count the total rows in df 

number_total = df.shape[0]
print(number_total)
```

```python
#step3,to culculate the proportion

proportion = carats / number_total
proportion
```

3. How many of the diamonds have a carat less than the mean carat value?

5. Which diamond has the highest price per carat? What is its value?

calculate the unit price of each diamond, then use max to select the highest value.


```python
#step1,get a new column called unitprice in the dataset

df['unitprice'] = df['price'] / df['carat']
df
```

```python
#step2, get the highest price of the diamond

highest_price = df[df['unitprice'] == df['unitprice'].max()]
highest_price
```

```python
#step3,
print('highest diamond price:', highest_price['unitprice'])
```
